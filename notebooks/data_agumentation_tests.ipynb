```python
import albumentations
import numpy as np
import matplotlib.pyplot as plt
```

```python
example = np.load("../data/MMs/Training/Labeled/A1E9Q1/A1E9Q1_sa_slice1_phase6.npy")
```

```python
plt.figure(figsize=(7,5))
plt.imshow(example, cmap="gray")
plt.axis("off")
```

### Flip

```python
aug = albumentations.Compose([
    albumentations.HorizontalFlip(p=1)
])
transformed = aug(image=example)["image"]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(17, 10))

ax1.imshow(example, cmap="gray")
ax1.axis("off")
ax1.set_title("Original")

ax2.imshow(transformed, cmap="gray")
ax2.axis("off")
ax2.set_title("Horizontal Flip")
```

```python
aug = albumentations.Compose([
    albumentations.VerticalFlip(p=1)
])
transformed = aug(image=example)["image"]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(17, 10))

ax1.imshow(example, cmap="gray")
ax1.axis("off")
ax1.set_title("Original")

ax2.imshow(transformed, cmap="gray")
ax2.axis("off")
ax2.set_title("Vertical Flip")
```

### Rotate

```python
aug = albumentations.Compose([
    albumentations.Rotate(p=1, limit=180, interpolation=1, border_mode=0)
])
transformed = aug(image=example)["image"]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(17, 10))

ax1.imshow(example, cmap="gray")
ax1.axis("off")
ax1.set_title("Original")

ax2.imshow(transformed, cmap="gray")
ax2.axis("off")
ax2.set_title("Rotate Limit 180")
```

### Shift

```python
aug = albumentations.Compose([
    albumentations.ShiftScaleRotate(p=1, shift_limit=0.25, scale_limit=0.0, rotate_limit=0, border_mode=0),  # shift
])
transformed = aug(image=example)["image"]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(17, 10))

ax1.imshow(example, cmap="gray")
ax1.axis("off")
ax1.set_title("Original")

ax2.imshow(transformed, cmap="gray")
ax2.axis("off")
ax2.set_title("Shift")
```

### Scale

```python
aug = albumentations.Compose([
    albumentations.ShiftScaleRotate(p=1, shift_limit=0.0, scale_limit=0.25, rotate_limit=0, border_mode=0),  # scale
])
transformed = aug(image=example)["image"]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(17, 10))

ax1.imshow(example, cmap="gray")
ax1.axis("off")
ax1.set_title("Original")

ax2.imshow(transformed, cmap="gray")
ax2.axis("off")
ax2.set_title("Scale")
```

### ElasticTransform

```python
aug = albumentations.Compose([
    albumentations.ElasticTransform(p=1, alpha=140, sigma=177 * 0.05, alpha_affine=100 * 0.03),
])
transformed = aug(image=example)["image"]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(17, 10))

ax1.imshow(example, cmap="gray")
ax1.axis("off")
ax1.set_title("Original")

ax2.imshow(transformed, cmap="gray")
ax2.axis("off")
ax2.set_title("Elastic Transform")
```

### Grid Distortion
